```python
# Let's create a function that displays the point line with the bar.
def visualize(theta, X, y):
    plt.scatter(X, y, color = 'blue')
    plt.plot(X, theta[0] * X + theta[1], color = 'blue')
    plt.show()
```

```python
# Create a function that multiplies each element of the matrix X by the slope of the model (theta[1]),
#followed by the addition of the intercept of the model (theta[0]), thus producing the predictions of the simple linear regression model.

def predict(X, theta):
    # return theta[0] + theta[1] * X
    return theta[0] * X + theta[1]
```

```python
def fit(X, y, theta, alpha, num_iters):
     # Initialize some useful variables
    m = X.shape[0]

    theta = np.zeros(2)

    # Unpack theta
    # a, b = theta

    # Loop over the number of iterations
    for _ in range(num_iters):
        # Compute the predictions
        predictions = predict(X, theta)

        # Compute the error
        error = predictions - y

        # Compute the gradient
        db = np.sum(error) / m
        da = np.sum(error * X) / m

        # Update theta
        theta[0] -= da * alpha
        theta[1] -= db * alpha
        # a -= da * alpha
        # b -= db * alpha
        # Perform one iteration of gradient descent (i.e., update theta once)
    return (theta)

# To begin, we'll set alpha to 0.01 and num_iters to 1000

# theta = np.random.rand(2) * 0.01
# (a, b) = fit(X, y, theta, 10e-3, 1000)

# print(a, b)

# def error(a, b, X, y):
#   predictions = predict(X, a, b)

#   # Compute the error
#   return (predictions - y) ** 2

# print(error(a, b, X, y).mean())
# visualize((b,a), X, y)

# print(error(21912.58918422329, 9880.814004608217, X, y).mean())
# visualize((21912.58918422329, 9880.814004608217), X, y)

#You should have a result similar to this one: [21912.58918422329, 9880.814004608217]
```
